```python
#IMPORT STATEMENTS

#IMPORTING NUMPY FOR WORKING WITH MATRICES
import numpy as np

#USING THE FACE_RECOGNITION LIBRARY AS OUR MAIN MODEL
import face_recognition

#USING OS FOR PATH RECOGNITION
import os

#IMPORTING CV2 FOR BOUNDING BOXES AND IMAGE PROCESSING

import cv2

#PANDAS TO GENERATE FINAL CSV
import pandas as pd
```

```python
#SPECIFYING PATH OF OUR DATA DIRECTORY WHERE TRAINING IMAGES ARE STORED
datadirectory= 'img'
path='img'

#MAKING 2 EMPTY LISTS TO ENCODE FACES AND THEIR NAMES FROM TRAINING IMAGES
knownfaces = []
knownnames = []

#FOR EVERY IMAGE IN THE DATASET, WE CALCULATE ITS ENCODING USING FACE_RECOGNTION LIBRARY
for name in os.listdir(datadirectory):
    
    image = face_recognition.load_image_file(path+"/"+name)
    
    #WE WANT ONLY 1 FACE FROM EACH TRAINING IMAGE, SO [0]
    
    encoding = face_recognition.face_encodings(image)[0]
    known_faces.append(encoding)
    known_names.append(name)
```

```python
#CALCULATING THE FRAMES PER SECOND OF OUR GIVEN FOOTAGE, TO MAINTAIN ORIGINAL FPS IN OUTPUT FILE

cam = cv2.VideoCapture('datavideo.mp4')
fps= int(cam.get(cv2.CAP_PROP_FPS))
print(fps)
```

```python
#EXTRACTING DATE OF CREATION OF VIDEO AS A METADATA, TO APPEND INTO CSV DATA FILE 

import datetime
import os

path = r"datavideo.mp4"
timestamp = os.path.getmtime(path)
datestamp = datetime.datetime.fromtimestamp(timestamp)
 
ctimestamp = os.path.getctime(path)
 
cdatestamp = datetime.datetime.fromtimestamp(ctimestamp)

d=str(cdatestamp)
d=d.split(" ")[0]
print(d)
```

```python
#CREATING AN OUTPUT VIDEO WITH SPECIFIED FRAMES PER SECOND, WHERE VIDEO IS OF 800X800 DIMENSION

videoname='outputvideofinalfps.avi'
video = cv2.VideoWriter(videoname, 0, fps, (800, 800))
```

```python
#CREATING 3 EMPTY LISTS TO APPEND THE DETECTED FACE AND ITS TIMESTAMP AND DATE FROM VIDEO
namesdetected=[]
frametime=[]
dates=[]
```

```python
#CREATING A CAMERA OBJECT TO CAPTURE DATA FROM OUR DATA VIDEO
cam = cv2.VideoCapture('datavideo.mp4')

imgtime=0
k1=1

#THE LOOP GETS IMAGES FROM VIDEO TILL THE VIDEO ENDS
while True:
    
    #CAPTURING A FRAME FROM THE VIDEO
    succ,image=cam.read()
    
    #IF FRAME WAS SUCCESSFULLY CAPTURED, ONLY THEN PROCESS
    if succ==True:
            
            #CV2 WORKS WITH BGR, SO CONVERSION OF FRAME FROM RBG TO BGR
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

            #EXTRACTING ALL FACE LOCATIONS FROM THE CAPTURED FRAME
            locations = face_recognition.face_locations(image)

            #CALCULATING ENCODING OF EACH DETECTED FACE
            encodings = face_recognition.face_encodings(image, locations)
            

            #ZIP IS USED TO ITERATE OVER EACH FACE AND ITS CORRESPONDING LOCATION
            for face_encoding, face_location in zip(encodings, locations):

                #MATCH STORES IF THE FACE MATCHED A KNOWN FACE OR NOT, USING COMPARE FACE FUNCTION OF THE LIBRARY      
                match = face_recognition.compare_faces(known_faces,face_encoding)
                
                #CALCULATINF DISTANCE OF THE FACE CAPTURED WITH EACH FACE IN THE KNOWN DATABASE
                faceDistance = face_recognition.face_distance(known_faces,face_encoding)
                
                #EXTRACTING THE FACE WITH THE MINIMUM DISTANCE/ CLOSEST MATCHING FACE
                matchIndex = np.argmin(faceDistance)
                
                #IF FACE HAS MATCHED AND ITS DISTANCE IS LESS THAN 0.515
                #HERE 0.515 IS A HYPER PARAMETER. IF FACE IS DETECTED, BUT ITS DISTANCE FROM ALL THE FACE IS MORE THAN
                #0.515, THEN THE FACE IS UNKNOWN
                if match[matchIndex] and faceDistance[matchIndex]<0.515:
                    
                    #FORMATTING THE NAME TO DISPLAY IN THE OUTPUT VIDEO 
                    names = known_names[matchIndex].upper()
                    names='person_'+names.split(".")[0]
                    
                    #IF THE FACE IS NEW, THEN APPEND ITS NAME, AND THE TIME STAMP AT WHICH IT OCCURS IN THE VIDEO
                    if names not in namesdetected:
                        namesdetected.append(names)
                        frametime.append(float(imgtime/fps))
                        dates.append(d)
                        
                    #EXTRACTING THE LOCATION OF THE FACE FOR THE BOUNDING BOX
                    y1,x2,y2,x1 = face_location
                    
                    #CREATING A BOUNDING BOX AROUND THE FACE
                    cv2.rectangle(image,(x1,y1),(x2,y2),(0,255,0),2)
                    cv2.rectangle(image,(x1,y2-35),(x2,y2),(0,255,0),cv2.FILLED)
                    cv2.putText(image,names,(x1+6,y2-6),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),2)
                    
                    
                    
                    
                    
                else :
                    
                    #IF THE FACE IS UNKNOWN, THEN EXTRACT ITS CO-ORDINATES
                    y1,x2,y2,x1 = face_location
                    
                    #BELOW COMMENTED CODE NAMES THE UNKNOWN FACE IMAGE TO STORE IT IN THE FOLDER
#                     nameofi="unknownimage"+str(k1)
#                     k1+=1

                    #CREATING A BOUNDING BOX AROUND THE UNKNOWN FACES
                    cv2.rectangle(image,(x1,y1),(x2,y2),(0,255,0),2)
                    cv2.rectangle(image,(x1,y2-35),(x2,y2),(0,255,0),cv2.FILLED)
                    cv2.putText(image,"UNKNOWN",(x1+6,y2-6),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),2)
                    
                    #THE BELOW COMMENTED CODE ALSO APPENDS THE FRAMES CONTAINING UNKNOWN FACES IN A SEPERATE FOLDER FOR FURTHER USE
#                     cv2.imwrite('unknowndata/'+nameofi+'.png',image)
                    

            
            imgtime+=1
            
            #RECOLOURING THE FRAME TO RGB AND WRITING IT TO OUR VIDEO FILE
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            image=cv2.resize(image,(800,800))
            video.write(image)
            
    else:
        
        #IF VIDEO ENDS, THE RELEASE OUR VIDEO OBJECT
        video.release()
        break
    
```

```python
#PRINTING ALL THE DETECTED FACES AND THEIR TIME STAMPS
print(namesdetected)
print(frametime)
#print(dates)
```

```python
#CAPTURING ONLY THE NAME OF THE DETECTED FACES
for i in range(len(namesdetected)):
    namesdetected[i]=namesdetected[i].split(".")[0]

  
```

```python
#CREATING A PANDAS DATAFRAME TO APPEND OUR NAMES,TIMESTAMPS AND DATES
df = pd.DataFrame({'Employee Name': namesdetected, 'Timestamp from footage(in seconds)': frametime,'DATES':dates}) 
```

```python
#CONVERTING THE DATAFRAME TO A CSV FILE
df.to_csv('record.csv', index=False)
```
